```python
#we want generated student data, [X1,...,X7] -> fundamentals, abstract, vision, intuition, difficulty]
import numpy as np
import pandas as pd
import random

# Creating a mock dataset
# Let's assume we have the following features: 'time_spent', 'performance_score', 'activity_count', 'content_difficulty'
# The target variable is 'content_appropriate' which is 1 if the content difficulty is right, and 0 if it's too difficult.

# Generate a mock dataset
np.random.seed(42)  # For reproducibility
def student_mappings(each_add):
    fundamentals = np.random.normal(.8,.1, size=each_add).tolist()
    fundamentals.extend(np.random.normal(.6,.15, size = each_add).tolist())
    fundamentals.extend(np.random.normal(.45,.12,size=each_add).tolist())
    print(len(fundamentals))

    abstract = np.random.normal(.45,.2, size=each_add).tolist()
    abstract.extend(np.random.normal(.3,.1,size=each_add).tolist())
    abstract.extend(np.random.normal(.1,.05,size=each_add).tolist())
    print(len(abstract))

    applied = np.random.normal(.45,.2, size=each_add).tolist()
    applied.extend(np.random.normal(.3,.1,size=each_add).tolist())
    applied.extend(np.random.normal(.1,.05,size=each_add).tolist())
    print(len(applied))

    vision = np.random.normal(.5,.15, size=each_add).tolist()
    vision.extend(np.random.normal(.3,.1,size=each_add).tolist())
    vision.extend(np.random.normal(.2,.05,size=each_add).tolist())
    print(len(vision))

    intuition = np.random.normal(.8,.1, size=each_add).tolist()
    intuition.extend(np.random.normal(.6,.15, size = each_add).tolist())
    intuition.extend(np.random.normal(.45,.12,size=each_add).tolist())
    print(len(intuition))

    user_difficulty = np.random.uniform(.3, .17, size = each_add).tolist()
    user_difficulty.extend(np.random.uniform(.45,.2, size = each_add).tolist())
    user_difficulty.extend(np.random.uniform(.7,.2,size=each_add).tolist())
    print(len(user_difficulty))
    # Combine into a DataFrame
    students = pd.DataFrame({
        'student_id': [i+1 for i in range(9999)],
        'fundamentals': fundamentals,
        'abstract': abstract,
        'applied': applied,
        'vision': vision,
        'user_difficulty': user_difficulty
    })

    return students


def question_to_student_mapping(each_add):
    fundamentals = np.random.normal(.8, .1, size=each_add).tolist()
    fundamentals.extend(np.random.normal(.6, .15, size=each_add).tolist())
    fundamentals.extend(np.random.normal(.45, .12, size=each_add).tolist())

    abstract = np.random.normal(.45, .2, size=each_add).tolist()
    abstract.extend(np.random.normal(.3, .1, size=each_add).tolist())
    abstract.extend(np.random.normal(.1, .05, size=each_add).tolist())

    applied = np.random.normal(.45, .2, size=each_add).tolist()
    applied.extend(np.random.normal(.3, .1, size=each_add).tolist())
    applied.extend(np.random.normal(.1, .05, size=each_add).tolist())

    vision = np.random.normal(.5, .15, size=each_add).tolist()
    vision.extend(np.random.normal(.3, .1, size=each_add).tolist())
    vision.extend(np.random.normal(.2, .05, size=each_add).tolist())
    # user_difficulty =

    intuition = np.random.normal(.8, .1, size=each_add).tolist()
    intuition.extend(np.random.normal(.6, .15, size=each_add).tolist())
    intuition.extend(np.random.normal(.45, .12, size=each_add).tolist())

    teacher_difficulty = np.random.uniform(.4, .12, size=each_add).tolist()
    teacher_difficulty.extend(np.random.uniform(.45, .2, size=each_add).tolist())
    teacher_difficulty.extend(np.random.uniform(.7, .2, size=each_add).tolist())

    questions = pd.DataFrame({
        'fundamentals': fundamentals,
        'abstract': abstract,
        'applied': applied,
        'vision': vision,
        'teacher_difficulty': teacher_difficulty
    })

    return questions

def responses_by_students(students, questions):
    question_response = pd.DataFrame()
    question_response['student_id'] = students['student_id']
    for i in range(len(questions)):
      prob_correct = []
      for j in range(len(students)):
            prob_correct.append(abs(students['fundamentals'][i] * students['abstract'][i] * students['applied'][i] * students['vision'][i] /students['user_difficulty'][i]))
            prob_correct[j] /= abs(questions['fundamentals'][i] * questions['abstract'][i] * questions['applied'][i] * questions['vision'][i] /questions['teacher_difficulty'][i])
      students_by_question = [prob_correct[i] for i in range(len(students))]

      col = f'question_{i}'
      question_response[col] = np.array(students_by_question)

    return question_response
```
